```python
vendas = pd.read_csv("/content/Base de Dados - Question치rio.xlsx - Sheet1.csv",decimal=',')
produtos = pd.read_csv("/content/Base de Dados - Question치rio.xlsx - Sheet2.csv")
consumidores = pd.read_csv("/content/Base de Dados - Question치rio.xlsx - Sheet3.csv")
lojas = pd.read_csv("/content/Base de Dados - Question치rio.xlsx - Sheet4.csv")
```

```python
vendaslimpo = vendas.drop(columns=['Unnamed: 0', 'Unnamed: 1'])
```

```python
vendaslimpo
```

```python
vendaslimpo = vendaslimpo.rename(columns=vendaslimpo.iloc[3])
```

```python
vendaslimpo = vendaslimpo.drop(index =[0,1,2,3])
```

```python
vendaslimpas = vendaslimpo.reset_index(drop = True)
```

```python
vendaslimpas
```

```python
vendaslimpas.head()
```

```python
vendaslimpas["Discount"]= vendaslimpas["Discount"].str.replace(',','.')
vendaslimpas["UnitPrice"]= vendaslimpas["UnitPrice"].str.replace(',','.')
```

```python
vendaslimpas = vendaslimpas.astype({'UnitPrice': float, 'Discount': float, 'Quantity' : int})
```

```python
vendaslimpas["DiscountValue"] = vendaslimpas["UnitPrice"] * vendaslimpas["Discount"]
```

```python
vendaslimpas
```

```python
vendaslimpas["SalePrice"] = (vendaslimpas["UnitPrice"] - vendaslimpas["DiscountValue"]) * vendaslimpas["Quantity"]
```

```python
vendaslimpas
```

```python
ReceitaLiquida = vendaslimpas[['SalePrice','Date']]
```

```python
ReceitaLiquida['Date'] = pd.to_datetime(ReceitaLiquida['Date'])
```

```python
ReceitaLiquida['Year'] = ReceitaLiquida['Date'].dt.year
```

```python
ReceitaLiquida2019 = ReceitaLiquida.drop('Date', axis = 1)
```
